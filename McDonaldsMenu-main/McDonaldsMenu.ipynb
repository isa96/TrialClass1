# Pertanyaan 3: Apakah grilled chicken yang dipesan menggantikan crispy chicken meningkatkan nilai gizi sandwich?

```python
#Mengambil data untuk sandwich
dataset_sandwich = dataset[dataset['Category']=='Chicken & Fish'][:4]
dataset_sandwich
```

```python
#Mengelompokkan antara crispy chicken dan grilled chicken
```

```python
dataset_sandwich_crispy_chicken = dataset_sandwich[dataset_sandwich['Item'].str.contains('Crispy')]
dataset_sandwich_crispy_chicken
```

```python
dataset_sandwich_grilled_chicken = dataset_sandwich[dataset_sandwich['Item'].str.contains('Grilled')]
dataset_sandwich_grilled_chicken
```

```python
#Menganalisa perbandingan sandwich crispy chicken dengan sandwich grilled chicken berdasarkan total fat (lemak total)
dataset_sandwich_fat = pd.DataFrame({'Grilled':
                                     [dataset_sandwich_grilled_chicken['Total Fat'].mean()],'Crispy':
                                    [dataset_sandwich_crispy_chicken['Total Fat'].mean()]})
dataset_sandwich_fat.plot.bar()
plt.ylabel('Total Fat')
plt.xlabel('Type of Sandwich')
plt.show()

print('Grilled Chicken memiliki lemak lebih tinggi dibandingkan dengan Crispy Chicken?'+
     str(dataset_sandwich_grilled_chicken['Total Fat'].mean()>dataset_sandwich_crispy_chicken['Total Fat'].mean()))
```

```python
#Menganalisa perbandingan sandwich crispy chicken dengan sandwich grilled chicken berdasarkan kalori
dataset_sandwich_cal = pd.DataFrame({'Grilled':
                                     [dataset_sandwich_grilled_chicken['Calories'].mean()],'Crispy':
                                    [dataset_sandwich_crispy_chicken['Calories'].mean()]})
dataset_sandwich_cal.plot.bar()
plt.ylabel('Calories')
plt.xlabel('Type of Sandwich')
plt.show()

print('Grilled Chicken memiliki kalori lebih tinggi dibandingkan dengan Crispy Chicken?'+
     str(dataset_sandwich_grilled_chicken['Calories'].mean()>dataset_sandwich_crispy_chicken['Calories'].mean()))
```

```python
#Menganalisa perbandingan sandwich crispy chicken dengan sandwich grilled chicken berdasarkan protein
dataset_sandwich_pro = pd.DataFrame({'Grilled':
                                     [dataset_sandwich_grilled_chicken['Protein'].mean()],'Crispy':
                                    [dataset_sandwich_crispy_chicken['Protein'].mean()]})
dataset_sandwich_pro.plot.bar()
plt.ylabel('Protein')
plt.xlabel('Type of Sandwich')
plt.show()

print('Grilled Chicken memiliki protein lebih tinggi dibandingkan dengan Crispy Chicken?'+
     str(dataset_sandwich_grilled_chicken['Protein'].mean()>dataset_sandwich_crispy_chicken['Protein'].mean()))
```

Kesimpulan untuk pertanyaan ketiga ini adalah bahwa dalam hal kalori dan total fat (lemak total), Crispy Chicken memiliki nilai yang lebih tinggi dibandingkan dengan Grilled Chicken, tetapi dari segi protein, Grilled Chicken lebih unggul dibandingkan dengan Crispy Chicken.

# Pertanyaan 4: Bagaimana dengan memesan putih telur daripada telur utuh?

```python
#Langkah 1 Mengambil data yang berhubungan dengan telur
dataset_egg = dataset[dataset['Item'].str.contains('Egg')]
dataset_egg
```

```python
#Langkah 2 mengambil data yang berhubungan dengan putih telur atau white egg
dataset_white_egg = dataset[dataset['Item'].str.contains('White')]
dataset_white_egg
```

```python
#Langkah 3 Mengambil data yang berhubungan dengan telur utuh atau whole egg
dataset_whole_egg = dataset_egg[~dataset['Item'].str.contains('White')]
dataset_whole_egg
```

```python
#Perbandingan kalori antara putih telur dengan telur utuh
dataset_egg_cal = pd.DataFrame({'White':
                               [dataset_white_egg['Calories'].mean()],
                               'Whole':
                               [dataset_whole_egg['Calories'].mean()]})
dataset_egg_cal.plot.bar()
plt.ylabel('Total Calorie')
plt.xlabel('Type of Egg')
plt.show()

print('Putih telur memiliki kandungan kalori yang lebih tinggi dibandingkan dengan telur utuh?'+
     str(dataset_white_egg['Calories'].mean()>dataset_whole_egg['Calories'].mean()))
```

```python
#Perbandingan protein antara putih telur dengan telur utuh
dataset_egg_cal = pd.DataFrame({'White':
                               [dataset_white_egg['Protein'].mean()],
                               'Whole':
                               [dataset_whole_egg['Protein'].mean()]})
dataset_egg_cal.plot.bar()
plt.ylabel('Total Protein')
plt.xlabel('Type of Egg')
plt.show()

print('Putih telur memiliki kandungan protein yang lebih tinggi dibandingkan dengan telur utuh?'+
     str(dataset_white_egg['Protein'].mean()>dataset_whole_egg['Protein'].mean()))
```

```python
#Perbandingan karbohidrat antara putih telur dengan telur utuh
dataset_egg_cal = pd.DataFrame({'White':
                               [dataset_white_egg['Carbohydrates'].mean()],
                               'Whole':
                               [dataset_whole_egg['Carbohydrates'].mean()]})
dataset_egg_cal.plot.bar()
plt.ylabel('Total Carbohydrate')
plt.xlabel('Type of Egg')
plt.show()

print('Putih telur memiliki kandungan karbohidrat yang lebih tinggi dibandingkan dengan telur utuh?'+
     str(dataset_white_egg['Carbohydrates'].mean()>dataset_whole_egg['Carbohydrates'].mean()))
```

```python
#Perbandingan lemak antara putih telur dengan telur utuh
dataset_egg_cal = pd.DataFrame({'White':
                               [dataset_white_egg['Total Fat'].mean()],
                               'Whole':
                               [dataset_whole_egg['Total Fat'].mean()]})
dataset_egg_cal.plot.bar()
plt.ylabel('Total Fat')
plt.xlabel('Type of Egg')
plt.show()

print('Putih telur memiliki kandungan lemak yang lebih tinggi dibandingkan dengan telur utuh?'+
     str(dataset_white_egg['Total Fat'].mean()>dataset_whole_egg['Total Fat'].mean()))
```

Untuk pertanyaan keempat dapat disimpulkan bahwa white egg (putih telur) memiliki kandungan kalori, protein, karbohidrat, dan lemak yang lebih banyak dibandingkan dengan whole egg (telur utuh).
